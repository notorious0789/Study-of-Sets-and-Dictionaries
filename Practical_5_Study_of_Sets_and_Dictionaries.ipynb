# SETS

```python
# Creation of Set
thisset = {"orange", "chili", "kiwi"}
print(thisset)
```

```python
# Unordered - No duplicate items
thisset = {"orange", "chili", "kiwi", "orange"}
print(thisset)
```

```python
# The Values True and 1 are considered the same value in sets, and are true
thisset = {"orange", "chili", "kiwi", True, 1, 2}
print(thisset)
```

```python
# The Values False and 0 are considered the same value in sets, and are false
thisset = {"orange", "chili", "kiwi", False, 0, 1}
print(thisset)
```

```python
# Once a set is created, you cannot replace its items, but you can add new items
thisset = {"orange", "chili", "kiwi"}
thisset.add("grapes")
print(thisset)
```

```python
# Remove Item
thisset = {"orange", "chili", "kiwi"}
thisset.remove("kiwi")
print(thisset)
```

```python
# Set Operations
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print("Union :", A | B)
print("Intersection :", A & B)
print("Difference :", A - B)
print("Symmetric difference :", A ^ B)
```

```python
# PROBLEM STATEMENT 2 - ELECTIVE SUBJECTS
'''Three students choose elective subjects.
Find the common subjects chosen by both students using set operations.'''

#Three students choose elective subjects. Find the common subjects chosen by both students using set operations.

Apoorba = ["AMT","EDP","PES","EG","Cybersecurity","VLSI"]
Mini = ["CLA","EDP","FEEE","EG","Cybersecurity","Embedded Systems"]
Anvi =  ["CLA","AIML","FEEE","EG","Cybersecurity","OOPS"]
common_subjects = set(Apoorba) & set(Mini) & set(Anvi)
print("Common Subjects:", common_subjects)
```

```python
# PROBLEM STATEMENT 3 - SPORTS CLUBS
'''One set contains students in the Cricket Club,another in the Football Club.
Find students who are :
1. In Both Clubs
2. In Only One of the Clubs.'''

# One set contains students in the Cricket Club,another in the Football Club. Find students who are : 1. In Both Clubs 2. In oOnly One of the Clubs.

cricket = {"Anvi","Rupsa","Arddra","Ansh","Apoorba"}
football = {"Rupsa","Apoorba","Arddra"}
print("Students in both clubs:", cricket & football)
print("Students in only one of the clubs:", cricket ^ football)
```

```python
# PROBLEM STATEMENT 4 - ABSENT STUDENTS
''' Given a set of all students and a set of present students,
find the absent students.'''
# Given a set of all students and a set of present students, find the absent students
students = {"Aarav", "Riya", "Kunal", "Pooja", "Sarthak", "Neha", "Rupsa"}
present_students = {"Aarav", "Rupsa", "Kunal", "Pooja"}
absent_students = students - present_students
print("The absent students are:")
print(absent_students)
```

```python
# PROBLEM STATEMENT 5 - REMOVE INVALID ENTRIES
# Remove invalid entries of courses

courses = ["FEEE", "EDP", "EG", "TPCS", "ECSF"]
print("Total courses:", courses)
actualcourses = set(courses)
actualcourses.remove("EG")
actualcourses.discard("ECSF")
print("Courses left:",actualcourses)
```

# DICTIONARIES


A Dictionary is a collecton which is ordered, changeable and do not allow duplicates

```python
# Creation of Dictionary
thisdict = {"Brand": "BMW", "Model": "M5 ", "Year": 2011 }
print(thisdict)
```

```python
# Dictionary takes latest value given for key. It will not allow duplicate values
thisdict = {"Brand": "BMW", "Model": "M5 ", "Year": 2011, "Year": 2020}
print(thisdict)
```

```python
# Length of Dictionary - Length of key-value pair
print(len(thisdict))
```

```python
# Changing /Adding key-value in dictionary
car = {"Brand": "BMW", "Model": "M5 ", "Year": 2011}
car["Year"] = 2020
print(car)
car["Color"] = "Black"
print(car)
x = car.values()
print(x) #after the change
```

```python
# Retrival of values
student = {"Roll No":101,"Name":"Ajay","Branch": "E&TC"}
print("Name:", student["Name"])
```

```python
# Adding and Updating Elements
student["Year"] = "Second"
student["Name"] = "Amit"
print(student)
```

```python
# Removing Elements
student.pop("Branch")
print("After Removal:",student)
```

```python
# PROBLEM STATEMENT 1 - PRODUCTS
''' 1. A Dictionary stores product names and their prices. WAPP to update the price of a given product.'''
#A Dictionary stores product names and their prices. WAPP to update the price of a given product.
products = {"Apple": 100, "Banana": 20, "Orange":50}
print("Original Products:",products)
products["Apple"] =55
print("Updated Products:",products)
```

```python
# PROBLEM STATEMENT 2 - STUDENT NAMES AND MARKS
''' 2. A Dictionary stores student names and their marks. WAPP to find the student's marks using the name entered. If the student is not found, display a suitable message.'''
marks = {"A":85, "B":90, "C":100}
name = input("Enter Student name:")
print(marks.get(name, "Student not Found"))
```

```python
# PROBLEM STATEMENT - 3 USERNAMES AND PASSWORDS
''' 3. A dictionary stores usernames and passwords. WAPP to validate user login'''
# A dictionary stores usernames and passwords. WAPP to validate user login
userdetails = {"admin":"1234", "Student":"abcd"}
user_name = input("Enter Username:")
password = input("Enter Password:")
if userdetails.get(user_name) == password:
    print("Login Successful")
else:
    print("Login Failed")
```

```python
# PROBLEM STATEMENT 4 - HIGHEST MARKS
'''4. A Dictionary stores student names and marks.
WAPP to find the student who scored the highest marks.'''
Details = {"Alan":85, "Blake":90, "Charlie":100}
topper = max(Details, key=Details.get)
print("Student with highest marks:", topper)
print("Marks:", Details[topper])
```
